```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.iloc[:, df.columns != 'Outcome']
y = df.iloc[:, df.columns == 'Outcome']
```

```python
print(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score # Import the accuracy_score function

# ... (rest of your code) ...

predict_output = model.predict(x_test) # Pass x_test to the predict method
print(predict_output)

acc = accuracy_score(y_test, predict_output) # Now accuracy_score is defined
print('the accuracy score for RF:', acc)
```
